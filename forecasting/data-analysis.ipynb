```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
i_path = 'D:\\Data\\Box-Office-Forecasting'
data = pd.read_csv(os.path.join(i_path, 'movie-master-final.csv'), header=0, sep=';', engine='python', encoding= 'utf8')
```

```python
data['year'] = data['year'].apply(lambda x: int(x))
```

```python
df = data.groupby('year')['is_profitable'].agg(lambda x: x.sum() / len(x)).reset_index(name='perc')
```

```python
file_name = "D:\\Master\\Thesis Related\\Yearly_Profitability_Percentages.xlsx"
df.to_excel(file_name, encoding= 'utf8', header=True, index=True)
```

```python
plt.plot( 'year', 'perc', data=df, color='skyblue')
plt.axhline(y=0.5, color='g', linestyle='-')
plt.rcParams["figure.figsize"] = (12, 6)
plt.xticks(np.arange(2000, 2019, step=2))
plt.show()
```
